```python
#USED
# requirements:
# pip install pillow tqdm scikit-image opencv-python --quiet

from pathlib import Path
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm

from scipy.signal import wiener
from skimage.morphology import disk, white_tophat
from skimage import exposure
import cv2
import os

# ---------------------------
# USER CONFIG
# ---------------------------
INPUT_DIR  = Path("/content/Dataset")   # change to your input folder
OUTPUT_DIR = Path("enhanced_dataset_all")  # root output folder
OUTPUT_DIR.mkdir(parents=True, exist_ok=True)

EXTS = (".png", ".jpg", ".jpeg", ".bmp", ".tif", ".tiff")

# Step subfolders (single folder per step)
STEP0_DIR = OUTPUT_DIR / "step0_foreground_rgba"      # raw RGBA from GrabCut
STEP1_DIR = OUTPUT_DIR / "fg_step1_gray"
STEP2_DIR = OUTPUT_DIR / "fg_step2_wiener"
STEP3_DIR = OUTPUT_DIR / "fg_step3_tophat"
STEP4_DIR = OUTPUT_DIR / "fg_step4_stretch"
STEP5_DIR = OUTPUT_DIR / "fg_step5_CLAHE"
STEP_FINAL_DIR = OUTPUT_DIR / "fg_final_rgba"         # final RGBA with processed grayscale as RGB

for d in (STEP0_DIR, STEP1_DIR, STEP2_DIR, STEP3_DIR, STEP4_DIR, STEP5_DIR, STEP_FINAL_DIR):
    d.mkdir(parents=True, exist_ok=True)

# ---------------------------
# HELPERS
# ---------------------------
def grayscale_custom(img_np):
    """Convert RGB numpy array to custom grayscale (uint8)."""
    R = img_np[..., 0].astype(float)
    G = img_np[..., 1].astype(float)
    B = img_np[..., 2].astype(float)
    gray = 0.2989 * R + 0.5870 * G + 0.1140 * B
    return np.clip(gray, 0, 255).astype(np.uint8)

def safe_name_from_path(p: Path, base: Path):
    rel = p.relative_to(base)
    rel_str = str(rel)
    safe = rel_str.replace(os.sep, "__").replace("/", "__").replace(" ", "_")
    return safe

def extract_foreground_grabcut(img_path, iter_count=5, rect_margin=10):
    """
    Use GrabCut to extract foreground.
    Returns:
      rgba: (H,W,4) uint8 with alpha 0 background, 255 foreground
      mask_fg: (H,W) uint8 mask where foreground==1, background==0
    """
    img_bgr = cv2.imread(str(img_path))
    if img_bgr is None:
        raise RuntimeError(f"cv2.imread failed for {img_path}")
    img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
    h, w = img_rgb.shape[:2]

    mask = np.zeros((h, w), np.uint8)
    bgdModel = np.zeros((1, 65), np.float64)
    fgdModel = np.zeros((1, 65), np.float64)

    x = rect_margin
    y = rect_margin
    rect_w = max(1, w - 2 * rect_margin)
    rect_h = max(1, h - 2 * rect_margin)
    rect = (x, y, rect_w, rect_h)

    cv2.grabCut(img_rgb, mask, rect, bgdModel, fgdModel, iter_count, cv2.GC_INIT_WITH_RECT)

    # foreground mask: sure or probable foreground
    mask_fg = np.where((mask == 1) | (mask == 3), 1, 0).astype('uint8')

    # apply mask to RGB image (background becomes 0)
    fg_rgb = img_rgb * mask_fg[:, :, np.newaxis]

    alpha = (mask_fg * 255).astype(np.uint8)
    rgba = np.dstack((fg_rgb, alpha))

    return rgba, mask_fg

def enhance_pipeline_on_foreground(rgb_fg, mask_fg):
    """
    rgb_fg: uint8 RGB image where background pixels are zero
    mask_fg: uint8 mask (1 foreground, 0 background)
    Returns: tuple of uint8 images (gray, wiener, tophat, stretched, clahe)
    All arrays are same HxW (single channel) and background remains 0.
    """
    # ensure float for computations where needed
    # 1) Grayscale from the foreground RGB (background black -> grayscale 0)
    gray = grayscale_custom(rgb_fg)

    # Force background to 0 explicitly (in case rgb_fg had non-zero due to compression)
    gray = np.where(mask_fg == 1, gray, 0).astype(np.uint8)

    # 2) Wiener filter (small neighborhood)
    denoised = wiener(gray.astype(float), (2, 2))
    denoised = np.clip(denoised, 0, 255).astype(np.uint8)
    denoised = np.where(mask_fg == 1, denoised, 0).astype(np.uint8)

    # 3) Top-hat filtering (illumination correction)
    selem = disk(15)
    tophat = white_tophat(denoised, selem)
    tophat = np.clip(tophat, 0, 255).astype(np.uint8)
    tophat = np.where(mask_fg == 1, tophat, 0).astype(np.uint8)

    # 4) Contrast stretching (percentile-based) -- compute percentiles only on foreground pixels
    fg_pixels = tophat[mask_fg == 1]
    if fg_pixels.size == 0:
        # nothing in foreground; return zeros
        stretched = np.zeros_like(tophat)
    else:
        p_low, p_high = np.percentile(fg_pixels, (15, 90))
        # if p_low == p_high, rescale_intensity would fail; handle it
        if p_low == p_high:
            stretched = tophat.copy()
        else:
            stretched = exposure.rescale_intensity(
                tophat, in_range=(p_low, p_high), out_range=(0, 255)
            ).astype(np.uint8)
    stretched = np.where(mask_fg == 1, stretched, 0).astype(np.uint8)

    # 5) CLAHE (applied only to foreground pixels)
    clahe = cv2.createCLAHE(clipLimit=1.2, tileGridSize=(1, 10))
    final_clahe = clahe.apply(stretched)
    final_clahe = np.where(mask_fg == 1, final_clahe, 0).astype(np.uint8)

    return gray, denoised, tophat, stretched, final_clahe

# ---------------------------
# COLLECT INPUT FILES
# ---------------------------
files = [p for p in INPUT_DIR.rglob("*") if p.suffix.lower() in EXTS]
files = sorted(files)
print(f"Found {len(files)} images in {INPUT_DIR}")

# ---------------------------
# PROCESS ALL IMAGES
# ---------------------------
preview_list = []
for idx, img_path in enumerate(tqdm(files, desc="Processing images")):
    safe_name = safe_name_from_path(img_path, INPUT_DIR)
    stem = Path(safe_name).stem

    # 0) Extract foreground (GrabCut)
    try:
        fg_rgba, mask_fg = extract_foreground_grabcut(img_path, iter_count=5, rect_margin=10)
    except Exception as e:
        print(f"Error extracting foreground for {img_path}: {e}")
        continue

    # save raw fg RGBA
    Image.fromarray(fg_rgba, mode="RGBA").save(STEP0_DIR / f"{stem}_fg_raw.png")

    # prepare rgb foreground (background = black)
    rgb_fg = fg_rgba[..., :3]

    # Run enhancement pipeline on foreground only
    try:
        gray, denoised, tophat, stretched, final_clahe = enhance_pipeline_on_foreground(rgb_fg, mask_fg)
    except Exception as e:
        print(f"Error enhancing foreground for {img_path}: {e}")
        continue

    # Save per-step foreground images (single-channel PNG)
    Image.fromarray(gray).save(STEP1_DIR / f"{stem}_fg_step1_gray.png")
    Image.fromarray(denoised).save(STEP2_DIR / f"{stem}_fg_step2_wiener.png")
    Image.fromarray(tophat).save(STEP3_DIR / f"{stem}_fg_step3_tophat.png")
    Image.fromarray(stretched).save(STEP4_DIR / f"{stem}_fg_step4_stretch.png")
    Image.fromarray(final_clahe).save(STEP5_DIR / f"{stem}_fg_step5_CLAHE.png")

    # Create final RGBA: use processed final_clahe as RGB channels (so the subject shows processed grayscale),
    # preserve the original alpha from GrabCut so the background remains transparent.
    rgb_processed = np.dstack((final_clahe, final_clahe, final_clahe)).astype(np.uint8)
    alpha = fg_rgba[..., 3]
    final_rgba = np.dstack((rgb_processed, alpha))
    Image.fromarray(final_rgba, mode="RGBA").save(STEP_FINAL_DIR / f"{stem}_fg_enhanced.png")

    # Collect previews for the first 5 images
    if idx < 5:
        preview_list.append((img_path.name, fg_rgba, mask_fg * 255, gray, denoised, tophat, stretched, final_clahe, final_rgba))

print("Processing complete.")
print("Saved results to:", OUTPUT_DIR)

# ---------------------------
# SHOW FIRST 5 RESULTS
# ---------------------------
for item in preview_list:
    (name, fg_rgba, mask_vis, gray, denoised, tophat, stretched, final_clahe, final_rgba) = item
    titles = ["Foreground RGBA", "Mask", "Gray", "Wiener", "Top-hat", "Stretch", "CLAHE", "Final RGBA"]
    images = [fg_rgba, mask_vis, gray, denoised, tophat, stretched, final_clahe, final_rgba]

    plt.figure(figsize=(18, 3))
    plt.suptitle(f"Preview: {name}", fontsize=14)

    for j, (title, img) in enumerate(zip(titles, images)):
        plt.subplot(1, len(images), j+1)
        if img is None:
            plt.text(0.5, 0.5, "N/A", ha="center", va="center")
            plt.title(title)
            plt.axis("off")
            continue

        if img.ndim == 2:
            plt.imshow(img, cmap="gray")
        else:
            # RGBA or RGB
            plt.imshow(img)
        plt.title(title)
        plt.axis("off")

    plt.show()
```

```python
import os
import cv2
import numpy as np

# 🔹 Normalization Function (same as before)
def normalize_fingerprint(img, desired_mean=100, desired_var=100):
    img = img.astype(np.float32)
    mean = np.mean(img)
    var = np.var(img)

    normalized = np.where(
        img > mean,
        desired_mean + np.sqrt((desired_var * (img - mean) ** 2) / (var + 1e-8)),
        desired_mean - np.sqrt((desired_var * (img - mean) ** 2) / (var + 1e-8))
    )

    return normalized


# 🔹 Input & Output folders
input_folder = "enhanced_dataset_all/fg_final_rgba"
output_folder = "enhanced_dataset_all/normalized_output"

os.makedirs(output_folder, exist_ok=True)

# 🔹 Process all images
for filename in os.listdir(input_folder):
    if filename.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp', '.tif')):
        img_path = os.path.join(input_folder, filename)

        # Read RGBA image
        img_rgba = cv2.imread(img_path, cv2.IMREAD_UNCHANGED)

        if img_rgba is None:
            continue

        # Convert RGBA → Grayscale
        if img_rgba.shape[2] == 4:
            img_gray = cv2.cvtColor(img_rgba, cv2.COLOR_BGRA2GRAY)
        else:
            img_gray = cv2.cvtColor(img_rgba, cv2.COLOR_BGR2GRAY)

        # Normalize fingerprint
        normalized = normalize_fingerprint(img_gray)

        # Scale to display range (0–255)
        normalized_display = cv2.normalize(normalized, None, 0, 255, cv2.NORM_MINMAX)
        normalized_display = normalized_display.astype(np.uint8)

        # Save result
        save_path = os.path.join(output_folder, filename)
        cv2.imwrite(save_path, normalized_display)

        print(f"Processed: {filename}")

print("✅ All fingerprints normalized and saved.")
```

```python
import os
import cv2
import numpy as np
from scipy.ndimage import gaussian_filter

# 🔹 Orientation Estimation Function
def estimate_orientation(normalized_img):
    sobelx = cv2.Sobel(normalized_img, cv2.CV_32F, 1, 0, ksize=3)
    sobely = cv2.Sobel(normalized_img, cv2.CV_32F, 0, 1, ksize=3)

    Gxx = sobelx ** 2
    Gyy = sobely ** 2
    Gxy = sobelx * sobely

    Gxx = gaussian_filter(Gxx, sigma=5)
    Gyy = gaussian_filter(Gyy, sigma=5)
    Gxy = gaussian_filter(Gxy, sigma=5)

    theta = 0.5 * np.arctan2(2 * Gxy, Gxx - Gyy + 1e-8)
    return theta


# 🔹 Folder Paths
input_folder = "enhanced_dataset_all/normalized_output"
output_folder = "enhanced_dataset_all/orientation_output"

os.makedirs(output_folder, exist_ok=True)

# 🔹 Process Each Image
for filename in os.listdir(input_folder):
    if filename.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp', '.tif')):
        img_path = os.path.join(input_folder, filename)

        # Read normalized grayscale image
        img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
        if img is None:
            continue

        img = img.astype(np.float32)

        # Estimate orientation
        theta = estimate_orientation(img)

        # 🔹 Save RAW orientation data (for later steps like Gabor)
        np.save(os.path.join(output_folder, filename + "_orientation.npy"), theta)

        # 🔹 Create visualization image (map angle → 0–255)
        vis = (theta + np.pi/2) / np.pi  # normalize to 0–1
        vis = (vis * 255).astype(np.uint8)

        cv2.imwrite(os.path.join(output_folder, filename), vis)

        print(f"Orientation computed for: {filename}")

print("✅ Orientation estimation completed for all images.")
```

```python
import os
import cv2
import numpy as np

def estimate_orientation_blockwise(img, block_size=16):
    img = img.astype(np.float32)

    gx = cv2.Sobel(img, cv2.CV_32F, 1, 0, ksize=3)
    gy = cv2.Sobel(img, cv2.CV_32F, 0, 1, ksize=3)

    rows, cols = img.shape
    orientation = np.zeros((rows // block_size, cols // block_size))

    for i in range(0, rows - block_size, block_size):
        for j in range(0, cols - block_size, block_size):
            block_gx = gx[i:i+block_size, j:j+block_size]
            block_gy = gy[i:i+block_size, j:j+block_size]

            Vx = 2 * np.sum(block_gx * block_gy)
            Vy = np.sum(block_gx**2 - block_gy**2)

            theta = 0.5 * np.arctan2(Vx, Vy + 1e-8)
            orientation[i//block_size, j//block_size] = theta

    return orientation
```

```python
def draw_orientation_field(img, orientation, block_size=16):
    vis = cv2.cvtColor(img.astype(np.uint8), cv2.COLOR_GRAY2BGR)

    rows, cols = orientation.shape

    for i in range(rows):
        for j in range(cols):
            theta = orientation[i, j]

            x = j * block_size + block_size // 2
            y = i * block_size + block_size // 2

            dx = int((block_size // 2) * np.cos(theta))
            dy = int((block_size // 2) * np.sin(theta))

            cv2.line(vis, (x - dx, y - dy), (x + dx, y + dy), (0, 0, 255), 1)

    return vis
```

```python
def compute_reliability(img, block_size=16):
    img = img.astype(np.float32)
    gx = cv2.Sobel(img, cv2.CV_32F, 1, 0, ksize=3)
    gy = cv2.Sobel(img, cv2.CV_32F, 0, 1, ksize=3)

    rows, cols = img.shape
    reliability = np.zeros((rows // block_size, cols // block_size))

    for i in range(0, rows - block_size, block_size):
        for j in range(0, cols - block_size, block_size):
            block_gx = gx[i:i+block_size, j:j+block_size]
            block_gy = gy[i:i+block_size, j:j+block_size]

            Gxx = np.sum(block_gx**2)
            Gyy = np.sum(block_gy**2)
            Gxy = np.sum(block_gx*block_gy)

            lambda1 = (Gxx + Gyy)/2 + np.sqrt(4*Gxy**2 + (Gxx - Gyy)**2)/2
            lambda2 = (Gxx + Gyy)/2 - np.sqrt(4*Gxy**2 + (Gxx - Gyy)**2)/2

            reliability[i//block_size, j//block_size] = 1 - (lambda2 / (lambda1 + 1e-8))

    return reliability
```

```python
import os
import cv2
import numpy as np

def compute_reliability_blockwise(img, block_size=16):
    img = img.astype(np.float32)
    gx = cv2.Sobel(img, cv2.CV_32F, 1, 0, ksize=3)
    gy = cv2.Sobel(img, cv2.CV_32F, 0, 1, ksize=3)

    rows, cols = img.shape
    reliability = np.zeros((rows // block_size, cols // block_size))

    for i in range(0, rows - block_size, block_size):
        for j in range(0, cols - block_size, block_size):
            block_gx = gx[i:i+block_size, j:j+block_size]
            block_gy = gy[i:i+block_size, j:j+block_size]

            Gxx = np.sum(block_gx**2)
            Gyy = np.sum(block_gy**2)
            Gxy = np.sum(block_gx * block_gy)

            lambda1 = (Gxx + Gyy)/2 + np.sqrt((Gxx - Gyy)**2 + 4*Gxy**2)/2
            lambda2 = (Gxx + Gyy)/2 - np.sqrt((Gxx - Gyy)**2 + 4*Gxy**2)/2

            reliability[i//block_size, j//block_size] = 1 - (lambda2 / (lambda1 + 1e-8))

    return reliability
```

```python
import cv2
import numpy as np

def rotate_block(block, angle):
    center = (block.shape[1]//2, block.shape[0]//2)
    rot_mat = cv2.getRotationMatrix2D(center, angle * 180/np.pi + 90, 1.0)
    rotated = cv2.warpAffine(block, rot_mat, (block.shape[1], block.shape[0]),
                             flags=cv2.INTER_LINEAR)
    return rotated
```

```python
def estimate_frequency_block(block, orientation):
    rotated = rotate_block(block, orientation)

    projection = np.sum(rotated, axis=0)
    projection = (projection - np.min(projection)) / (np.max(projection) - np.min(projection) + 1e-8)

    peaks = []
    for i in range(1, len(projection)-1):
        if projection[i] > projection[i-1] and projection[i] > projection[i+1]:
            peaks.append(i)

    if len(peaks) < 2:
        return 0

    distances = np.diff(peaks)
    avg_spacing = np.mean(distances)

    if avg_spacing == 0:
        return 0

    return 1.0 / avg_spacing
```

```python
def compute_ridge_frequency(img, orientation, reliability, block_size=16):
    rows, cols = img.shape
    freq_map = np.zeros((rows // block_size, cols // block_size))

    for i in range(0, rows - block_size, block_size):
        for j in range(0, cols - block_size, block_size):

            r = i // block_size
            c = j // block_size

            if reliability[r, c] < 0.5:   # skip unreliable blocks
                continue

            block = img[i:i+block_size, j:j+block_size]
            theta = orientation[r, c]

            freq = estimate_frequency_block(block, theta)
            freq_map[r, c] = freq

    return freq_map
```

```python
import os

norm_folder = "enhanced_dataset_all/normalized_output"
ori_folder = "enhanced_dataset_all/orientation_data"
rel_folder = "enhanced_dataset_all/reliability_data"
freq_folder = "enhanced_dataset_all/frequency_data"

os.makedirs(freq_folder, exist_ok=True)

for filename in os.listdir(norm_folder):
    if filename.lower().endswith(('.png','.jpg','.jpeg','.bmp','.tif')):

        img = cv2.imread(os.path.join(norm_folder, filename), cv2.IMREAD_GRAYSCALE).astype(np.float32)
        orientation = np.load(os.path.join(ori_folder, filename + "_orientation.npy"))
        reliability = np.load(os.path.join(rel_folder, filename + "_reliability.npy"))

        freq_map = compute_ridge_frequency(img, orientation, reliability)

        np.save(os.path.join(freq_folder, filename + "_freq.npy"), freq_map)

        vis = (freq_map / (np.max(freq_map)+1e-8) * 255).astype(np.uint8)
        cv2.imwrite(os.path.join(freq_folder, filename), vis)

        print("Frequency computed:", filename)
```

```python
def estimate_frequency_block(block, orientation):
    rotated = rotate_block(block, orientation)

    projection = np.sum(rotated, axis=0)
    projection = (projection - np.min(projection)) / (np.max(projection) - np.min(projection) + 1e-8)

    peaks = []
    for i in range(1, len(projection)-1):
        if projection[i] > projection[i-1] and projection[i] > projection[i+1]:
            peaks.append(i)

    if len(peaks) < 2:
        return 0

    distances = np.diff(peaks)
    avg_spacing = np.mean(distances)

    if avg_spacing == 0:
        return 0

    return 1.0 / avg_spacing
```

```python
def compute_ridge_frequency(img, orientation, reliability, block_size=16):
    rows, cols = img.shape
    freq_map = np.zeros((rows // block_size, cols // block_size))

    for i in range(0, rows - block_size, block_size):
        for j in range(0, cols - block_size, block_size):

            r = i // block_size
            c = j // block_size

            if reliability[r, c] < 0.5:  # skip unreliable regions
                continue

            block = img[i:i+block_size, j:j+block_size]
            theta = orientation[r, c]

            freq = estimate_frequency_block(block, theta)
            freq_map[r, c] = freq

    return freq_map
```

```python
def visualize_frequency_map(freq_map, img_shape):
    freq_norm = freq_map / (np.max(freq_map) + 1e-8)

    vis = cv2.resize(freq_norm, (img_shape[1], img_shape[0]),
                     interpolation=cv2.INTER_NEAREST)

    vis = (vis * 255).astype(np.uint8)
    return vis
```

```python
import os
import cv2
import numpy as np
from tqdm import tqdm
```

```python
def gabor_kernel(theta, freq, ksize=21):
    sigma_x = 1.0 / freq
    sigma_y = 1.0 / freq

    half = ksize // 2
    y, x = np.mgrid[-half:half+1, -half:half+1]

    x_theta = x * np.cos(theta) + y * np.sin(theta)
    y_theta = -x * np.sin(theta) + y * np.cos(theta)

    kernel = np.exp(
        -0.5 * ((x_theta**2) / sigma_x**2 +
                (y_theta**2) / sigma_y**2)
    ) * np.cos(2 * np.pi * freq * x_theta)

    return kernel
```

```python
def gabor_enhancement(norm_img, orientation, frequency, reliability,
                      block_size=16, rel_thresh=0.5):

    h, w = norm_img.shape
    enhanced = np.zeros_like(norm_img, dtype=np.float32)

    for i in range(0, h - block_size, block_size):
        for j in range(0, w - block_size, block_size):

            ci = i + block_size // 2
            cj = j + block_size // 2

            # reliability mask (paper requirement)
            if reliability[ci, cj] < rel_thresh:
                continue

            freq = frequency[ci, cj]
            if freq <= 0:
                continue

            theta = orientation[ci, cj]

            kernel = gabor_kernel(theta, freq)

            block = norm_img[i:i+block_size, j:j+block_size]
            filtered = cv2.filter2D(block, cv2.CV_32F, kernel)

            enhanced[i:i+block_size, j:j+block_size] = filtered

    return enhanced
```

```python
def run_gabor_on_enhanced_dataset(root_dir):

    norm_dir = os.path.join(root_dir, "normalized_output")
    orient_dir = os.path.join(root_dir, "orientation_data")
    freq_dir = os.path.join(root_dir, "frequency_data")
    rel_dir = os.path.join(root_dir, "reliability_data")

    out_dir = os.path.join(root_dir, "gabor_output")
    os.makedirs(out_dir, exist_ok=True)

    image_names = sorted([
        f for f in os.listdir(norm_dir)
        if f.lower().endswith((".png", ".jpg", ".bmp"))
    ])

    for name in tqdm(image_names, desc="Gabor filtering"):

        # ---- load inputs ----
        norm = cv2.imread(os.path.join(norm_dir, name),
                          cv2.IMREAD_GRAYSCALE).astype(np.float32)

        orient = np.load(os.path.join(
            orient_dir, name.replace(".png", ".npy")))

        freq = np.load(os.path.join(
            freq_dir, name.replace(".png", ".npy")))

        rel = np.load(os.path.join(
            rel_dir, name.replace(".png", ".npy")))

        # ---- gabor filtering ----
        enhanced = gabor_enhancement(norm, orient, freq, rel)

        # ---- mask low reliability ----
        enhanced[rel < 0.5] = 0

        # ---- normalize & save ----
        enhanced = cv2.normalize(
            enhanced, None, 0, 255, cv2.NORM_MINMAX
        ).astype(np.uint8)

        cv2.imwrite(os.path.join(out_dir, name), enhanced)
```

```python
import os
import cv2
import numpy as np
from scipy.ndimage import gaussian_filter
from skimage.filters import gabor_kernel
from skimage.morphology import remove_small_objects, remove_small_holes

# ==============================
# STEP 1 — FOREGROUND MASK (FIXED)
# ==============================
def create_foreground_mask(img):
    # Convert float image → 8-bit for Otsu
    img_uint8 = cv2.normalize(img, None, 0, 255, cv2.NORM_MINMAX).astype(np.uint8)

    blur = cv2.GaussianBlur(img_uint8, (15, 15), 0)

    _, mask = cv2.threshold(blur, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

    mask = cv2.morphologyEx(mask, cv2.MORPH_CLOSE, np.ones((15, 15), np.uint8))

    return mask.astype(np.float32) / 255.0


# ==============================
# STEP 2 — ORIENTATION ESTIMATION
# ==============================
def estimate_orientation(img):
    sobelx = cv2.Sobel(img, cv2.CV_32F, 1, 0, ksize=3)
    sobely = cv2.Sobel(img, cv2.CV_32F, 0, 1, ksize=3)

    Gxx = gaussian_filter(sobelx**2, sigma=5)
    Gyy = gaussian_filter(sobely**2, sigma=5)
    Gxy = gaussian_filter(sobelx * sobely, sigma=5)

    theta = 0.5 * np.arctan2(2 * Gxy, Gxx - Gyy + 1e-8)
    reliability = np.sqrt((Gxx - Gyy)**2 + 4 * Gxy**2) / (Gxx + Gyy + 1e-8)

    return theta, reliability


# ==============================
# STEP 3 — RIDGE FREQUENCY (SIMPLIFIED)
# ==============================
def estimate_frequency(img, block_size=32):
    rows, cols = img.shape
    freq = np.full_like(img, 0.1)  # Typical fingerprint ridge frequency
    return freq


# ==============================
# STEP 4 — GABOR ENHANCEMENT
# ==============================
def gabor_enhance(img, orientation, frequency, reliability):
    enhanced = np.zeros_like(img)

    block = 16
    for i in range(0, img.shape[0] - block, block):
        for j in range(0, img.shape[1] - block, block):
            theta = orientation[i, j]
            freq = frequency[i, j]

            if freq <= 0:
                continue

            kernel = np.real(gabor_kernel(frequency=freq, theta=theta))
            filtered = cv2.filter2D(img, cv2.CV_32F, kernel)

            enhanced[i:i+block, j:j+block] = filtered[i:i+block, j:j+block]

    return enhanced * reliability


# ==============================
# STEP 5 — POSTPROCESS (FINAL OUTPUT)
# ==============================
def postprocess(enhanced, original):
    mask = create_foreground_mask(original)

    enhanced = cv2.normalize(enhanced, None, 0, 255, cv2.NORM_MINMAX).astype(np.uint8)

    # Improve contrast
    enhanced = cv2.equalizeHist(enhanced)

    # Binarize ridges
    binary = cv2.adaptiveThreshold(
        enhanced, 255,
        cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
        cv2.THRESH_BINARY_INV,
        15, 5
    )

    # Remove noise
    binary = remove_small_objects(binary.astype(bool), min_size=100)
    binary = remove_small_holes(binary, area_threshold=100)
    binary = (binary.astype(np.uint8) * 255)

    # Apply fingerprint mask
    binary = binary * mask.astype(np.uint8)

    return binary


# ==============================
# MAIN PIPELINE
# ==============================
input_folder = "enhanced_dataset_all/normalized_output"
output_folder = "enhanced_dataset_all/final_enhanced_output"
os.makedirs(output_folder, exist_ok=True)

for filename in os.listdir(input_folder):
    if filename.lower().endswith(('.png', '.jpg', '.bmp', '.tif')):
        print("Processing:", filename)

        img = cv2.imread(os.path.join(input_folder, filename), cv2.IMREAD_GRAYSCALE)
        img = img.astype(np.float32) / 255.0

        orientation, reliability = estimate_orientation(img)
        frequency = estimate_frequency(img)
        enhanced = gabor_enhance(img, orientation, frequency, reliability)
        final_img = postprocess(enhanced, img)

        cv2.imwrite(os.path.join(output_folder, filename), final_img)

print("✅ Done! Check: enhanced_dataset_all/final_enhanced_output")
```

```python
#2
import os
import cv2
import numpy as np
from scipy.ndimage import gaussian_filter

# ========== ORIENTATION FUNCTION (FROM PAPER) ==========
def estimate_orientation(img, block_size=16):
    gx = cv2.Sobel(img, cv2.CV_32F, 1, 0, ksize=3)
    gy = cv2.Sobel(img, cv2.CV_32F, 0, 1, ksize=3)

    gxx = gx * gx
    gyy = gy * gy
    gxy = gx * gy

    # Gaussian weighting sigma = 10 (paper)
    gxx = gaussian_filter(gxx, sigma=10)
    gyy = gaussian_filter(gyy, sigma=10)
    gxy = gaussian_filter(gxy, sigma=10)

    rows, cols = img.shape
    orient_rows = rows // block_size
    orient_cols = cols // block_size

    orientation = np.zeros((orient_rows, orient_cols), dtype=np.float32)

    for i in range(orient_rows):
        for j in range(orient_cols):
            y0, y1 = i*block_size, (i+1)*block_size
            x0, x1 = j*block_size, (j+1)*block_size

            Gxx = np.sum(gxx[y0:y1, x0:x1])
            Gyy = np.sum(gyy[y0:y1, x0:x1])
            Gxy = np.sum(gxy[y0:y1, x0:x1])

            theta = 0.5 * np.arctan2(2*Gxy, Gxx - Gyy + 1e-8)
            orientation[i, j] = theta

    # Convert to continuous vector field
    phi_x = np.cos(2 * orientation)
    phi_y = np.sin(2 * orientation)

    # Smooth vector field sigma = 5 (paper)
    phi_x = gaussian_filter(phi_x, sigma=5)
    phi_y = gaussian_filter(phi_y, sigma=5)

    smooth_orientation = 0.5 * np.arctan2(phi_y, phi_x)

    return smooth_orientation
```

```python
import os
import cv2
import numpy as np
from scipy.signal import find_peaks

block_size = 16
peak_window = 9
min_wave = 1
max_wave = 50

norm_folder = "enhanced_dataset_all/normalized_output"
orient_folder = "enhanced_dataset_all/orientation_data"
freq_folder = "enhanced_dataset_all/frequency_data"

os.makedirs(freq_folder, exist_ok=True)


def compute_frequency_block(block, theta):
    """Compute ridge frequency for one block using x-signature"""
    rows, cols = block.shape
    center_i, center_j = rows // 2, cols // 2

    cos_t = np.cos(theta)
    sin_t = np.sin(theta)

    x_signature = []

    # Equation (2),(3),(4) from paper
    for k in range(block_size):
        vals = []
        for d in range(block_size):
            u = int(center_i + (d - block_size/2) * cos_t + (k - block_size/2) * sin_t)
            v = int(center_j + (d - block_size/2) * sin_t + (block_size/2 - k) * cos_t)

            if 0 <= u < rows and 0 <= v < cols:
                vals.append(block[u, v])

        if len(vals) > 0:
            x_signature.append(np.mean(vals))
        else:
            x_signature.append(0)

    x_signature = np.array(x_signature)

    # Smooth the signal
    x_signature = cv2.GaussianBlur(x_signature.reshape(-1,1), (peak_window,1), 0).flatten()

    # Find peaks
    peaks, _ = find_peaks(x_signature)

    if len(peaks) < 2:
        return 0  # Cannot compute frequency

    # Compute distances between consecutive peaks
    distances = np.diff(peaks)
    T = np.mean(distances)

    if T < min_wave or T > max_wave:
        return 0

    return 1.0 / T


# Process all images
for fname in os.listdir(norm_folder):
    if not fname.lower().endswith(".png"):
        continue

    print("Processing frequency:", fname)

    img = cv2.imread(os.path.join(norm_folder, fname), cv2.IMREAD_GRAYSCALE).astype(np.float32)
    orientation = np.load(os.path.join(orient_folder, fname.replace(".png", ".png_orientation.npy")))

    rows, cols = img.shape
    freq_map = np.zeros_like(orientation)

    for i in range(freq_map.shape[0]):
        for j in range(freq_map.shape[1]):
            r = i * block_size
            c = j * block_size

            block = img[r:r+block_size, c:c+block_size]
            theta = orientation[i, j]

            if block.shape != (block_size, block_size):
                continue

            freq_map[i, j] = compute_frequency_block(block, theta)

    np.save(os.path.join(freq_folder, fname.replace(".png", ".png_freq.npy")), freq_map)

print("✅ Ridge frequency estimation complete")
```

```python
import cv2
import numpy as np
import os
from scipy.ndimage import gaussian_filter

def compute_reliability(img, block_size=16):
    gx = cv2.Sobel(img, cv2.CV_32F, 1, 0, ksize=3)
    gy = cv2.Sobel(img, cv2.CV_32F, 0, 1, ksize=3)

    gxx = gx * gx
    gyy = gy * gy
    gxy = gx * gy

    # Same Gaussian sigma as orientation step (paper)
    gxx = gaussian_filter(gxx, sigma=10)
    gyy = gaussian_filter(gyy, sigma=10)
    gxy = gaussian_filter(gxy, sigma=10)

    rows, cols = img.shape
    r_rows = rows // block_size
    r_cols = cols // block_size

    reliability = np.zeros((r_rows, r_cols), dtype=np.float32)

    for i in range(r_rows):
        for j in range(r_cols):
            y0, y1 = i*block_size, (i+1)*block_size
            x0, x1 = j*block_size, (j+1)*block_size

            Gxx = np.sum(gxx[y0:y1, x0:x1])
            Gyy = np.sum(gyy[y0:y1, x0:x1])
            Gxy = np.sum(gxy[y0:y1, x0:x1])

            # Eigenvalues of structure tensor
            common = np.sqrt((Gxx - Gyy)**2 + 4*(Gxy**2))
            zeta_max = (Gxx + Gyy + common) / 2
            zeta_min = (Gxx + Gyy - common) / 2

            # Reliability from paper
            reliability[i, j] = 1 - (zeta_min / (zeta_max + 1e-8))

    # Normalize to 0–1
    reliability = np.clip(reliability, 0, 1)

    return reliability
```

```python
def gabor_enhancement(img, orientation, frequency, block=16):
    rows, cols = img.shape
    enhanced = np.zeros((rows, cols), dtype=np.float32)

    ksize = 21
    sigma = 4.0
    gamma = 0.5

    nby, nbx = orientation.shape  # number of blocks

    for by in range(nby):
        for bx in range(nbx):
            theta = orientation[by, bx]
            freq  = frequency[by, bx]

            if freq <= 0:
                continue

            y = by * block
            x = bx * block

            if y + block > rows or x + block > cols:
                continue

            kernel = cv2.getGaborKernel(
                (ksize, ksize),
                sigma,
                theta,
                1.0 / freq,
                gamma,
                0,
                ktype=cv2.CV_32F
            )

            patch = img[y:y+block, x:x+block]
            filtered = cv2.filter2D(patch, cv2.CV_32F, kernel)

            enhanced[y:y+block, x:x+block] = filtered

    return enhanced
```

```python
def image_to_block_map(img, block):
    h, w = img.shape
    bh, bw = h // block, w // block
    block_map = np.zeros((bh, bw), dtype=np.float32)

    for by in range(bh):
        for bx in range(bw):
            patch = img[
                by*block:(by+1)*block,
                bx*block:(bx+1)*block
            ]
            block_map[by, bx] = np.mean(patch)

    return block_map
```

```python
def run_gabor_filtering(base_dir, block=16):
    norm_dir = os.path.join(base_dir, "normalized_output")
    ori_dir  = os.path.join(base_dir, "orientation_data")
    freq_dir = os.path.join(base_dir, "frequency_data")
    rel_dir  = os.path.join(base_dir, "reliability_data")

    out_dir  = os.path.join(base_dir, "gabor_output")
    os.makedirs(out_dir, exist_ok=True)

    images = [f for f in os.listdir(norm_dir) if f.endswith(".png")]

    for img_name in tqdm(images, desc="Gabor filtering"):
        img = cv2.imread(
            os.path.join(norm_dir, img_name),
            cv2.IMREAD_GRAYSCALE
        )

        orientation = np.load(
            os.path.join(ori_dir, img_name + "_orientation.npy")
        )

        freq_img = cv2.imread(
            os.path.join(freq_dir, img_name),
            cv2.IMREAD_GRAYSCALE
        ).astype(np.float32) / 255.0

        rel_img = cv2.imread(
            os.path.join(rel_dir, img_name),
            cv2.IMREAD_GRAYSCALE
        ).astype(np.float32) / 255.0

        frequency   = image_to_block_map(freq_img, block)
        reliability = image_to_block_map(rel_img, block)

        enhanced = gabor_enhancement(img, orientation, frequency, block)

        enhanced = cv2.normalize(enhanced, None, 0, 255, cv2.NORM_MINMAX)
        enhanced = enhanced.astype(np.uint8)

        # Reliability masking (>50%)
        for by in range(reliability.shape[0]):
            for bx in range(reliability.shape[1]):
                if reliability[by, bx] < 0.5:
                    y = by * block
                    x = bx * block
                    enhanced[y:y+block, x:x+block] = 0

        _, binary = cv2.threshold(
            enhanced, 0, 255,
            cv2.THRESH_BINARY + cv2.THRESH_OTSU
        )

        cv2.imwrite(os.path.join(out_dir, img_name), binary)

    print("✅ Gabor filtering completed successfully")
```

```python
def gabor_filter_blockwise(img, orientation, frequency, block=16):
    h, w = img.shape
    out = np.zeros_like(img, dtype=np.float32)

    bh, bw = orientation.shape

    for by in range(bh):
        for bx in range(bw):
            freq = frequency[by, bx]
            theta = orientation[by, bx]

            if freq <= 1e-6:
                continue

            y = by * block
            x = bx * block

            kernel = cv2.getGaborKernel(
                (21, 21),
                sigma=4.0,
                theta=theta,
                lambd=1.0 / freq,
                gamma=0.5,
                psi=0,
                ktype=cv2.CV_32F
            )

            patch = img[y:y+block, x:x+block]
            if patch.shape != (block, block):
                continue

            filtered = cv2.filter2D(patch, cv2.CV_32F, kernel)
            out[y:y+block, x:x+block] = filtered

    return cv2.normalize(out, None, 0, 255, cv2.NORM_MINMAX).astype(np.uint8)
```

```python
def reliability_mask(binary, reliability, block=16):
    bh, bw = reliability.shape

    for by in range(bh):
        for bx in range(bw):
            if reliability[by, bx] < 0.5:
                y = by * block
                x = bx * block
                binary[y:y+block, x:x+block] = 0

    return binary
```

```python
def run_gabor_pipeline(base_dir):
    norm_dir = os.path.join(base_dir, "normalized_output")
    ori_dir  = os.path.join(base_dir, "orientation_data")
    freq_dir = os.path.join(base_dir, "frequency_data")
    rel_dir  = os.path.join(base_dir, "reliability_data")
    out_dir  = os.path.join(base_dir, "gabor_output")

    os.makedirs(out_dir, exist_ok=True)

    images = sorted([f for f in os.listdir(norm_dir) if f.endswith(".png")])

    for name in images:
        img = cv2.imread(os.path.join(norm_dir, name), 0)

        orientation = np.load(
            os.path.join(ori_dir, name + "_orientation.npy")
        )

        freq_img = cv2.imread(os.path.join(freq_dir, name), 0)
        rel_img  = cv2.imread(os.path.join(rel_dir, name), 0)

        frequency = image_to_block_map(freq_img) / 255.0
        reliability = image_to_block_map(rel_img) / 255.0

        gabor_soft = gabor_filter_blockwise(img, orientation, frequency)

        _, binary = cv2.threshold(
            gabor_soft, 0, 255,
            cv2.THRESH_BINARY + cv2.THRESH_OTSU
        )

        final = reliability_mask(binary, reliability)

        out_name = name.replace(".png", "_gabor_binary.png")
        cv2.imwrite(os.path.join(out_dir, out_name), final)
```

```python
def normalize_frequency(freq_img):
    freq = freq_img.astype(np.float32)

    # Remove background zeros
    freq[freq < 10] = 0

    # Normalize to ridge frequency range
    # Typical ridge spacing: 8–16 pixels
    freq = cv2.normalize(freq, None, 1/16, 1/8, cv2.NORM_MINMAX)

    return freq
```

```python
def normalize_reliability(rel_img):
    rel = rel_img.astype(np.float32) / 255.0
    return rel
```

```python
def gabor_soft_blockwise(img, orientation, frequency, block=16):
    h, w = img.shape
    out = np.zeros((h, w), np.float32)

    bh, bw = orientation.shape

    for by in range(bh):
        for bx in range(bw):
            freq = frequency[by, bx]
            theta = orientation[by, bx]

            if freq <= 0:
                continue

            y = by * block
            x = bx * block

            kernel = cv2.getGaborKernel(
                (21, 21),
                sigma=4.0,
                theta=theta,
                lambd=1.0 / freq,
                gamma=0.5,
                psi=0,
                ktype=cv2.CV_32F
            )

            patch = img[y:y+block, x:x+block]
            if patch.shape != (block, block):
                continue

            out[y:y+block, x:x+block] = cv2.filter2D(
                patch, cv2.CV_32F, kernel
            )

    return cv2.normalize(out, None, 0, 255, cv2.NORM_MINMAX).astype(np.uint8)
```

```python
def mask_by_reliability(soft, reliability, block=16):
    masked = soft.copy()
    bh, bw = reliability.shape

    for by in range(bh):
        for bx in range(bw):
            if reliability[by, bx] < 0.5:
                y = by * block
                x = bx * block
                masked[y:y+block, x:x+block] = 0

    return masked
```

```python
def binarize_soft(img):
    return cv2.adaptiveThreshold(
        img, 255,
        cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
        cv2.THRESH_BINARY,
        15, 2
    )
```

```python
def run_gabor_pipeline(base_dir):
    norm_dir = base_dir + "/normalized_output"
    ori_dir  = base_dir + "/orientation_data"
    freq_dir = base_dir + "/frequency_data"
    rel_dir  = base_dir + "/reliability_data"
    out_dir  = base_dir + "/gabor_output"

    os.makedirs(out_dir, exist_ok=True)

    images = [f for f in os.listdir(norm_dir) if f.endswith(".png")]

    for name in images:
        img = cv2.imread(f"{norm_dir}/{name}", 0)
        orientation = np.load(f"{ori_dir}/{name}_orientation.npy")

        freq_img = cv2.imread(f"{freq_dir}/{name}", 0)
        rel_img  = cv2.imread(f"{rel_dir}/{name}", 0)

        frequency   = image_to_block_map(normalize_frequency(freq_img))
        reliability = image_to_block_map(normalize_reliability(rel_img))

        soft = gabor_soft_blockwise(img, orientation, frequency)
        masked = mask_by_reliability(soft, reliability)
        binary = binarize_soft(masked)

        cv2.imwrite(
            f"{out_dir}/{name.replace('.png','_gabor.png')}",
            binary
        )
```
